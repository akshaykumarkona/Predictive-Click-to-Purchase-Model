## Data Pre-processing

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python
s=ss()
ag=s.fit_transform(ds[['Age']])
ag
```

```python
ds['SAge']=ag
ds
```

```python
es=s.fit_transform(ds[['EstimatedSalary']])
es
```

```python
ds['Ssalary']=es
ds
```

```python
ds.drop(['Age','EstimatedSalary'],axis=1,inplace=True)
ds
```

```python
import seaborn as sns
sns.heatmap(ds.corr(),annot=True)
plt.figure(figsize=(50,30))
```

```python
X=ds[['SAge','Ssalary']]
Y=ds['Purchased']
print(X,Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,ytrain,ytest=tts(X,Y,test_size=0.25,random_state=50)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Model Building and Evaluation

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import metrics
```

```python
accuracy=[]
for i in range(1,101):
    knc=KNeighborsClassifier(n_neighbors=i)
    knc.fit(X_train,ytrain)
    pred=knc.predict(X_test)
    accuracy.append(metrics.accuracy_score(ytest,pred))
print(accuracy)
```

## Selecting "k" value for KNN 

```python
plt.plot(accuracy)
custom_xticks = list(i for i in range(0,101,5))
plt.xticks(custom_xticks)
plt.figure(figsize=(100, 10)) 
plt.show()
```

```python
new_knc=KNeighborsClassifier(n_neighbors=8)
new_knc.fit(X_train,ytrain)
new_pred=new_knc.predict(X_test)
```

```python
metrics.confusion_matrix(ytest,new_pred)
```

```python
acc=metrics.accuracy_score(ytest,new_pred)
acc
```

```python
print('Therefore, the accuracy of this KNN Classifier is: ',acc*100,"%") 
```

### Accuracy of this KNN Classifier model is: 95.0 %
